```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Reading in the CSV for the monthly PE ratios of the S&P 500
SPdf = pd.read_csv("C:\\Users\\Philip\\Documents\\SP500_PE_RATIO_MONTH.csv")
```

```python
SPdf.head()
```

```python
#Permanently flip the data into chronological order instead of descending
SPdf = SPdf[::-1]

#Only include the last 100 years
SPdf = SPdf[(SPdf['Date'] >= '1921-01-01')]
SPdf.head()
```

```python
# NEW COLUMN called 'Change' which shows change in PE value
SPdf['Change'] = SPdf['Value'].diff()
```

```python
#Plotting the CHANGE in PE Ratio
SPdf.plot(x = 'Date', y = 'Change', title = 'Change in PE')
```

```python
#1929 CRASH (Change in PE)
SPdf[(SPdf['Date'] > '1926-01-01') & (SPdf['Date'] < '1931-01-01')].plot(x = 'Date', y = 'Change', kind = 'line', title = '1929 (Great Depression)')
```

```python
#1989 CRASH (Change in PE)
SPdf[(SPdf['Date'] > '1984-01-01') & (SPdf['Date'] < '1991-01-01')].plot(x = 'Date', y = 'Change', kind = 'line', title = '1989 (Black Monday)')
```

```python
#2001 Dot Com Bubble CRASH (Change in PE)
SPdf[(SPdf['Date'] > '1999-01-01') & (SPdf['Date'] < '2008-01-01')].plot(x = 'Date', y = 'Change', kind = 'line', title = '2001 (Dot Com Bubble)')
```

```python
#2008 The Great Recession (Change in PE)
SPdf[(SPdf['Date'] > '2005-01-01') & (SPdf['Date'] < '2012-01-01')].plot(x = 'Date', y = 'Change', kind = 'line', title = '2008 (The Great Recession)')
```

```python
#PE Ratio over the last 5 years
SPdf[(SPdf['Date'] > '2016-01-01') & (SPdf['Date'] < '2021-04-01')].plot(x = 'Date', y = 'Change', kind = 'line', title = '2016 to 2021')
```
